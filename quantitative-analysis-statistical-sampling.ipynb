Amostra aleatória simples

```python
amostra_simples = df.sample(n=50)
print(amostra_simples)
```

```python
observacoes = amostra_simples.shape[0]
print("Número de observações: ", observacoes)
```

```python
proporcao = (amostra_simples['Storage_GB_SSD'] >= 256).mean()
print("Proporção : ", proporcao)
```

```python
desvio = amostra_simples['Storage_GB_SSD'].std()
print("Desvio padrão: ", desvio)
```

Amostra estratificada por RAM_GB

```python
proporcoes_screen = df['Screen'].value_counts(normalize=True)
n_por_grupo = (proporcoes_screen * 50).round().astype(int)
print(n_por_grupo)
```

```python
amostra_estratificada = df.groupby('Screen', group_keys=False).apply(lambda x: x.sample(n=n_por_grupo[x.name], random_state=42))
print(amostra_estratificada)
```

```python
observacoes = amostra_estratificada.shape[0]
print("Número de observações: ", observacoes)
```

```python
proporcao = (amostra_estratificada['Storage_GB_SSD'] >= 256).mean()
print("Proporção : ", proporcao)
```

```python
desvio = amostra_estratificada['Storage_GB_SSD'].std()
print("Desvio padrão: ", desvio)
```
