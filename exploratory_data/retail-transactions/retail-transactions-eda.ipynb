**Passo 6** - **Remover** linhas duplicadas

```python
data_tran_1.drop_duplicates(inplace = True)
```

**Passo 7** - **Normalizar** coluna tran_date para o formato DD/MM/AAAA

```python
data_tran_1['tran_date'] = pd.to_datetime(data_tran_1['tran_date']).dt.strftime('%d/%m/%Y')
```

**Passo 8** - Converter coluna tran_date para o tipologia **datetime**

```python
data_tran_1['tran_date'] = pd.to_datetime(data_tran_1['tran_date'], dayfirst=True)
```

**Passo 9** - Criar **dicionário** chamado dicio para usar na **normalização** da coluna Store_type

```python
dicio = {
    'Flagship store': 'Flagship',
    'e-Shop': 'Online',
    'TeleShop': 'Telefone',
    'MBR': 'Presencial'
}
```

**Passo 10** - **Normalizar** coluna Store_type usando o dicio

```python
data_tran_1['Store_type'] = data_tran_1['Store_type'].map(dicio).fillna(data_tran_1['Store_type'])
```
