```python
# Criando o schema bronze

spark.sql("""
CREATE SCHEMA IF NOT EXISTS bronze          
""")
```

```python
from pyspark.sql.functions import current_timestamp

# Importar dados brutos de todos os CSVs

# 1. CO2 Emissions
df_co2_raw = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load("/Workspace/Users/sguimaraesa@gmail.com/exercício_final_ big_ data/dataset/co2_emissions_yearly.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
)

print(f"✅ CO2 Emissions carregado: {df_co2_raw.count()} registros")
```

```python
# 2. Energy Mix
df_energy_raw = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load("/Workspace/Users/sguimaraesa@gmail.com/exercício_final_ big_ data/dataset/energy_mix_yearly.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
)

print(f"✅ Energy Mix carregado: {df_energy_raw.count()} registros")
```

```python
# 3. Temperature Anomaly
df_temp_raw = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load("/Workspace/Users/sguimaraesa@gmail.com/exercício_final_ big_ data/dataset/temperature_anomaly_monthly.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
)

print(f"✅ Temperature Anomaly carregado: {df_temp_raw.count()} registros")
```

```python
# 4. Climate Events
df_climate_raw = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load("/Workspace/Users/sguimaraesa@gmail.com/exercício_final_ big_ data/dataset/climate_events.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
)

print(f"✅ Climate Events carregado: {df_climate_raw.count()} registros")
```

```python
# 5. Carbon Prices
df_carbon_raw = (
    spark.read
    .format("csv")
    .option("header", "true")
    .option("inferSchema", "true")
    .load("/Workspace/Users/sguimaraesa@gmail.com/exercício_final_ big_ data/dataset/carbon_prices_daily.csv")
    .withColumn("ingestion_timestamp", current_timestamp())
)

print(f"✅ Carbon Prices carregado: {df_carbon_raw.count()} registros")
```

```python
# Salvar na Bronze Layer

df_co2_raw.write.format("delta").mode("overwrite").saveAsTable("bronze.co2_emissions_raw")
df_energy_raw.write.format("delta").mode("overwrite").saveAsTable("bronze.energy_mix_raw")
df_temp_raw.write.format("delta").mode("overwrite").saveAsTable("bronze.temperature_anomaly_raw")
df_climate_raw.write.format("delta").mode("overwrite").saveAsTable("bronze.climate_events_raw")
df_carbon_raw.write.format("delta").mode("overwrite").saveAsTable("bronze.carbon_prices_raw")

print("✅ Todas as tabelas Bronze criadas com sucesso!")
```
