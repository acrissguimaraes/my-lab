**Passo 3** - *Identificar missing values* e converte-los para o **formato NaN**.

```python
Carros2.isnull().sum()
```

```python
import numpy as np
```

```python
Carros2.replace('?',np.nan, inplace = True)
```

**Passo 4** - Corrigir **tipologia** das colunas, caso necessário.

```python
Carros2["height"] = Carros2["height"].astype("float")
Carros2["horsepower"] = Carros2["horsepower"].astype("float")
Carros2["price"] = Carros2["price"].astype("float")

print(Carros2.dtypes)
```

**Passo 5** - Indicar o **total** de observações com *missing values* e o **número de *missing values* por variável**.

```python
print("Missing values por variável:")
print(Carros2.isnull().sum())

print("-" * 30)

total_linhas_com_na = Carros2.isnull().any(axis=1).sum()

print(f"Total de observações com missing values: {total_linhas_com_na}")
```

**Passo 6** - No caso da variável **price**, caso ela tenha missing values, faz **drop** das observações.

```python
Carros2.dropna(subset=["price"], axis=0, inplace=True)
Carros2.reset_index(drop=True, inplace=True)
```

**Passo 7** - Caso existam missing values noutras **variáveis quantitativas**, substitui-os pela **média**.

```python
avg_hp = Carros2["horsepower"].astype("float").mean(axis=0)
Carros2["horsepower"] = Carros2["horsepower"].replace(np.nan, avg_hp)

avg_height = Carros2["height"].astype("float").mean(axis=0)
Carros2["height"] = Carros2["height"].replace(np.nan, avg_height)
```

**Passo 8** - Caso existam *missing values* noutras **variáveis qualitativas**, substitui-os pela **moda**.

```python
most_common_fuel = Carros2['fuel-type'].value_counts().idxmax()
Carros2["fuel-type"] = Carros2["fuel-type"].replace(np.nan, most_common_fuel)

most_common_doors = Carros2['num-of-doors'].value_counts().idxmax()
Carros2["num-of-doors"] = Carros2["num-of-doors"].replace(np.nan, most_common_doors)
```

**Passo 9** - Confirma que não existem mais valores em falta no dataframe e que está **pronto para a fase seguinte**.

```python
print(Carros2.isnull().sum())
```

**Passo 10** - Avaliar os **principais indicadores** das variáveis quantitativas e verifica se existe um **problema de escalas**.

```python
print(Carros2.describe())
```

**Passo 11** - Criar uma nova variável **length_min_max** através da transformação **Min-Max** aplicada à variável **length**. Valida se os valores da nova variável são os **previstos pela transformação**.

```python
Carros2['length_min_max'] = (Carros2['length'] - Carros2['length'].min()) / (Carros2['length'].max() - Carros2['length'].min())

print(Carros2[['length', 'length_min_max']].describe())
```

**Passo 12** - Criar uma nova variável **width_std** através da **estandardização** da variável **width**. Validar se os novos valores da variável são os **previstos pela transformação**.

```python
Carros2['width_std'] = (Carros2['width'] - Carros2['width'].mean()) / Carros2['width'].std()
print(Carros2[['width', 'width_std']].describe())
```

**Passo 13** - Criar uma nova variável **height_log** através de uma transformação **Logarítmica** aplicada a **height**. Validar se os novos valores da variável são os **previstos pela transformação**.

```python
Carros2['height_log'] = np.log(Carros2['height'])
print(Carros2[['height', 'height_log']].head())
```

**Passo 14** - Cria uma nova variável **price_decile** através da transformação **em decis** da variável **price**. Valida se os valores da nova variável são os **previstos pela transformação**.

```python
Carros2['price_decile'] = pd.qcut(Carros2['price'], q=10, labels=False) + 1
print(Carros2['price_decile'].value_counts().sort_index())
```
